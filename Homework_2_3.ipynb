Joint probability distributions

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example: Joint Probability Distribution for two random variables X and Y
# Assume both X and Y follow a normal distribution

# Set mean and covariance matrix for the multivariate normal distribution
mean = [0, 0]  # Mean for X and Y
covariance = [[1, 0.5], [0.5, 1]]  # Covariance matrix

# Generate 1000 samples for X and Y
X, Y = np.random.multivariate_normal(mean, covariance, 1000).T

# Create a 2D histogram to represent the joint distribution
plt.hist2d(X, Y, bins=30, cmap='Blues')
plt.colorbar(label='Counts')

plt.title('2D Histogram of Joint Distribution')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()

# For 3D visualization of the joint probability distribution
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Compute the histogram
hist, xedges, yedges = np.histogram2d(X, Y, bins=30)

# Construct arrays for the positions of the bars.
xpos, ypos = np.meshgrid(xedges[:-1] + 0.25, yedges[:-1] + 0.25, indexing="ij")
xpos = xpos.ravel()
ypos = ypos.ravel()
zpos = 0

# Construct arrays with the dimensions for the 3D bars.
dx = dy = 0.5 * np.ones_like(zpos)
dz = hist.ravel()

# Create the 3D bar graph
ax.bar3d(xpos, ypos, zpos, dx, dy, dz, zsort='average')

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Frequency')
plt.title('3D Visualization of Joint Probability Distribution')

plt.show()
```

Correlation and dependence

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr, spearmanr

# Generate two sets of data for correlation and dependence
np.random.seed(42)

# Linear relationship (strong correlation)
X_linear = np.random.normal(0, 1, 1000)
Y_linear = 2 * X_linear + np.random.normal(0, 1, 1000)

# Non-linear relationship (dependence without correlation)
X_nonlinear = np.random.uniform(-2, 2, 1000)
Y_nonlinear = X_nonlinear**2 + np.random.normal(0, 0.1, 1000)

# No relationship (no correlation or dependence)
X_random = np.random.normal(0, 1, 1000)
Y_random = np.random.normal(0, 1, 1000)

# Pearson Correlation (Linear)
pearson_corr_linear, _ = pearsonr(X_linear, Y_linear)
pearson_corr_random, _ = pearsonr(X_random, Y_random)

# Spearman Correlation (for non-linear relationships)
spearman_corr_nonlinear, _ = spearmanr(X_nonlinear, Y_nonlinear)

print(f"Pearson Correlation (Linear relationship): {pearson_corr_linear:.2f}")
print(f"Pearson Correlation (Random relationship): {pearson_corr_random:.2f}")
print(f"Spearman Correlation (Non-linear relationship): {spearman_corr_nonlinear:.2f}")

# Plot the relationships to visualize correlation and dependence

plt.figure(figsize=(15, 5))

# 1. Linear relationship (Strong correlation)
plt.subplot(1, 3, 1)
plt.scatter(X_linear, Y_linear, alpha=0.5)
plt.title(f'Linear Relationship\nPearson Correlation: {pearson_corr_linear:.2f}')
plt.xlabel('X')
plt.ylabel('Y')

# 2. Non-linear relationship (Dependence without correlation)
plt.subplot(1, 3, 2)
plt.scatter(X_nonlinear, Y_nonlinear, alpha=0.5)
plt.title(f'Non-linear Relationship\nSpearman Correlation: {spearman_corr_nonlinear:.2f}')
plt.xlabel('X')
plt.ylabel('Y')

# 3. No relationship (No correlation)
plt.subplot(1, 3, 3)
plt.scatter(X_random, Y_random, alpha=0.5)
plt.title(f'No Relationship\nPearson Correlation: {pearson_corr_random:.2f}')
plt.xlabel('X')
plt.ylabel('Y')

plt.tight_layout()
plt.show()
```

2.3.3. Random samples

```python
import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# 1. Uniform distribution (between 0 and 1)
uniform_samples = np.random.uniform(0, 1, 1000)

# 2. Normal (Gaussian) distribution (mean=0, std=1)
normal_samples = np.random.normal(0, 1, 1000)

# 3. Binomial distribution (n=10, p=0.5)
binomial_samples = np.random.binomial(n=10, p=0.5, size=1000)

# 4. Exponential distribution (lambda=1)
exponential_samples = np.random.exponential(scale=1, size=1000)

# Plotting the random samples from different distributions

plt.figure(figsize=(12, 8))

# 1. Uniform distribution
plt.subplot(2, 2, 1)
plt.hist(uniform_samples, bins=30, color='skyblue', edgecolor='black')
plt.title('Uniform Distribution (0, 1)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 2. Normal distribution
plt.subplot(2, 2, 2)
plt.hist(normal_samples, bins=30, color='lightgreen', edgecolor='black')
plt.title('Normal Distribution (mean=0, std=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 3. Binomial distribution
plt.subplot(2, 2, 3)
plt.hist(binomial_samples, bins=30, color='orange', edgecolor='black')
plt.title('Binomial Distribution (n=10, p=0.5)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# 4. Exponential distribution
plt.subplot(2, 2, 4)
plt.hist(exponential_samples, bins=30, color='lightcoral', edgecolor='black')
plt.title('Exponential Distribution (=1)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
