2.4.1. Maximum Likelihood Estimation (MLE) for Random Samples


```python
import numpy as np
from scipy.optimize import minimize

# Generate random samples from an exponential distribution
np.random.seed(42)
data = np.random.exponential(scale=2, size=1000)  # True scale (1/θ) is 2

# Define the negative log-likelihood function
def neg_log_likelihood(theta, data):
    if theta <= 0:
        return np.inf  # To ensure we don't estimate non-positive θ
    n = len(data)
    return -n * np.log(theta) + theta * np.sum(data)

# Perform MLE by minimizing the negative log-likelihood
result = minimize(neg_log_likelihood, x0=1, args=(data,))
theta_mle = result.x[0]

print(f"MLE for θ: {theta_mle}")
```

2.4.2. Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Generating random data for the example
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)  # y = 4 + 3x + noise

# Fit linear regression model
model = LinearRegression()
model.fit(X, y)
y_pred = model.predict(X)

# Display the coefficients
print(f"Intercept (β0): {model.intercept_[0]}")
print(f"Slope (β1): {model.coef_[0][0]}")

# Plot the results
plt.scatter(X, y, color="blue", label="Data points")
plt.plot(X, y_pred, color="red", label="Fitted line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```
